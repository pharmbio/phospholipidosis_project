```python
import pandas as pd
```

```python
grit2 = pd.read_csv('data/specs2k_grit_viability.csv')
grit3 = pd.read_csv('data/specs3k_grit_viability.csv')
```

```python
plp = pd.read_csv('data/24h_fixed-Table 1.csv',sep=';')
```

```python
plp['Batch_nr'] = plp['Batch_nr'].str.replace('DMSO','PHB000001')
plp
```

```python
plp = plp.drop(columns=['Plate', 'Well', 'Concentration uM'])
```

```python
#plp['%Induction'].str.split(' ')[0]#.replace(',','.', regex=True).astype(float)
plp['Cell_to_DMSO']=plp['Cell_to_DMSO'].str.replace('%','').str.replace(',','.').astype(float)
plp['%Induction']=plp['%Induction'].str.replace(',','.').astype(float)

#plp['Cell_to_DMSO'] = plp['Cell_to_DMSO'].str.split(' ')[0]#.replace(',','.', regex=True).astype(float)
#plp
```

```python
# group by Batch_nr
plp = plp.groupby(['Batch_nr', "Compound_ID"]).agg('mean').reset_index()
columns_to_round = [
    "High_PLP_10uM_>80",
    "No_PLP_10uM_<4.53",
    "Low_PLP_10uM_between",
    "nontoxic_10uM_>80",
    "confirmed_in_fix"
]

# Round the selected columns and convert to int
plp[columns_to_round] = plp[columns_to_round].round().astype(int)

plp
```
